**1.** Construya una funci칩n que reciba como par치metros el ROE (en decimales), el plowback
(en decimales), y el capital inicial y que retorne:
a. Un DataFrame con 500 filas donde cada corresponde a un a침o. Incluya las
siguientes columnas: Capital Inicial, utilidades, dividendos, acumulaci칩n de capital
y capital final.
b. La tasa a la cual crecen los dividendos.

```python
#Importacion de la libreria Pandas
import pandas as pd

#Declaracion de variables y asignacion de valores correspondientes (seran parametros de la funcion)
ROE = 0.25
PLOWBACK = 0.60
Cap_inicial = 30
Tasa_descuento = 0.20
```

```python
# Definicion de la funci칩n que retornara el Dataframe

def DDM_Function(Return_equity, P_back, Capital_inicial):

      # Definicion e inicializacion de variables usadas para los calculos
      plazo = 500
      datos = [] # Lista que sera llenada con listas para asignar al DataFrame
      a침os = 0
      Capital_i = Capital_inicial
      Capital_f = 0
      Roe = Return_equity
      Utilidades = 0
      Plowback = P_back
      Dividendo = 0
      Acum_Capital = 0

      # Iteracion sobre los a침os desde 1 hasta 'plazo' (es decir 500)
      for i in range (1, plazo+1):
          a침os = i
          Utilidades = Capital_i * Roe
          Dividendo = Utilidades * (1-Plowback)
          Acum_Capital = Utilidades - Dividendo
          Capital_f = Capital_i + Acum_Capital

      # Se agrega la lista con los datos del periodo 'i' a la lista 'Datos'
          datos.append([a침os, Capital_i, Utilidades, Dividendo, Acum_Capital, Capital_f ])
          Capital_i = Capital_f #Actualizacion de la variable


      # Se crea un DataFrame a partir de la lista 'datos' con columnas solicitadas y se agregan como filas 'datos'
      data = pd.DataFrame(datos,
                          columns = ["A침os","Capital_Inicial", "Utilidades", "Dividendos", "Acumulacion_Capital", "Capital_final"]
                         )

      # Se cambia el indice de default a que los a침os (1-500) sean el indice
      data = data.set_index("A침os")

      # Se crea una nueva columna que contendra la tasa de crecimiento de los dividendos
      data["Dividiendos_RateGrowth"] = data.Dividendos.pct_change()  # Funcion pct_change() fue sacada del libro "Phyton para finanzas" de Carlos Mario Ramirez Gil
                                                                     # Ramirez Gil, C. M. (2021). Obteniendo y procesando datos financieros.
                                                                     # En A. Guitierrez (Ed.), Python para Finanzas (pp. 121-122). Editorial Ra-Ma.
      # Retorno del DataFrame
      return data
```

```python
# LLamado (o invocacion) de la funcion

calculo = DDM_Function(ROE, PLOWBACK, Cap_inicial)
calculo
```

```python
# Definicion de la funcion que retornara el valor de la accion

def stock_value(value, Discount_rate):

    # Inicializacion de la columna (se realizo con apoyo de ChatGPT)
    value["Div_Descontado"] = 0

    # Preguntar al profesor por el plazo - Iteracion sobre todos los plazos (1-500) para llenar la columna de Dividendos descontados
    for i in range(1, 501):
        value.loc[i,"Div_Descontado"] = 1 / (((1+Discount_rate)**i))* value.loc[i, "Dividendos"]
    precio = value.loc[:,"Div_Descontado"].sum()  # Funcion sum() realizada con apoyo de material de la clase
    return precio
```
