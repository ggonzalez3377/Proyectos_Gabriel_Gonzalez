# **Valoracion de una firma**

Un inversionista está considerando invertir en una firma cuyo ROE es 25% y la tasa plowback es del
𝑏 = 60%. Hoy en día, la firma cuenta con un capital de treinta dólares por acción 𝐾0 = $30. El
ROE y el plowback son constantes. La tasa de capitalización de mercado para descontar los
dividendos es 20%.





**1.** Función que recibe como parámetros el ROE (en decimales), el plowback
(en decimales), y el capital inicial y que retorna:
a. Un DataFrame con 500 filas donde cada corresponde a un año. Incluya las
siguientes columnas: Capital Inicial, utilidades, dividendos, acumulación de capital
y capital final.
b. La tasa a la cual crecen los dividendos.

```python
#Importacion de la libreria Pandas
import pandas as pd

#Declaracion de variables y asignacion de valores correspondientes (seran parametros de la funcion)
ROE = 0.25
PLOWBACK = 0.60
Cap_inicial = 30
Tasa_descuento = 0.20
```

```python
# Definicion de la función que retornara el Dataframe

def DDM_Function(Return_equity, P_back, Capital_inicial):

      # Definicion e inicializacion de variables usadas para los calculos
      plazo = 500
      datos = [] # Lista que sera llenada con listas para asignar al DataFrame
      años = 0
      Capital_i = Capital_inicial
      Capital_f = 0
      Roe = Return_equity
      Utilidades = 0
      Plowback = P_back
      Dividendo = 0
      Acum_Capital = 0

      # Iteracion sobre los años desde 1 hasta 'plazo' (es decir 500)
      for i in range (1, plazo+1):
          años = i
          Utilidades = Capital_i * Roe
          Dividendo = Utilidades * (1-Plowback)
          Acum_Capital = Utilidades - Dividendo
          Capital_f = Capital_i + Acum_Capital

      # Se agrega la lista con los datos del periodo 'i' a la lista 'Datos'
          datos.append([años, Capital_i, Utilidades, Dividendo, Acum_Capital, Capital_f ])
          Capital_i = Capital_f #Actualizacion de la variable


      # Se crea un DataFrame a partir de la lista 'datos' con columnas solicitadas y se agregan como filas 'datos'
      data = pd.DataFrame(datos,
                          columns = ["Años","Capital_Inicial", "Utilidades", "Dividendos", "Acumulacion_Capital", "Capital_final"]
                         )

      # Se cambia el indice de default a que los años (1-500) sean el indice
      data = data.set_index("Años")

      # Se crea una nueva columna que contendra la tasa de crecimiento de los dividendos
      data["Dividiendos_RateGrowth"] = data.Dividendos.pct_change()  
      # Retorno del DataFrame
      return data
```

```python
# LLamado (o invocacion) de la funcion

calculo = DDM_Function(ROE, PLOWBACK, Cap_inicial)
calculo
```

**2.** Función que recibe como parámetros el DataFrame del numeral anterior
junto con una tasa de descuento (en decimales) y retorna el valor de una acción. 

```python
# Definicion de la funcion que retornara el valor de la accion

def stock_value(value, Discount_rate):

    value["Div_Descontado"] = 0
    # Iteracion sobre todos los plazos (1-500) para llenar la columna de Dividendos descontados
    for i in range(1, 501):
        value.loc[i,"Div_Descontado"] = 1 / (((1+Discount_rate)**i))* value.loc[i, "Dividendos"]
    precio = value.loc[:,"Div_Descontado"].sum()  # Funcion sum() 
    return precio
```

```python
# Llamado (o invocacion) de la funcion
valor_accion = stock_value(calculo, Tasa_descuento)

print('El valor de la acciones es: ${:,}'.format(round(valor_accion)))
```
